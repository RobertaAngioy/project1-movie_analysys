MOVIE ANALYSIS 

Overview

Data Preparation

Join and clean data.

'Rating' and 'Title' have a column called 'tconst' that uniquely identifies each movie. I used it to join the two dataframes, and, to increase clarity, I renamed the resulting dataframe. I then dropped the null data.

```python
#Join ratings and title

movies=ratings.join(title.set_index('tconst'), on='tconst')
movies.head()
```

```python
#Â check nulls rows

movies.isnull().sum()
```

```python
# Drop null rows

movies.dropna(inplace=True)
movies.info()
movies
```

I renamed the column containing the title in the 'Movies' dataframe. I used the renamed column to join 'Movies' with the 'Gross' dataframe.

```python
#Rename title column

movies=movies.rename(columns={'primary_title':'title'})
movies.head()
```

```python
# Join movies with gross

movies = movies.join(gross.set_index('title'), on='title') 
movies.info()
movies.head()

```

The dataframe called 'Budget' required some steps before joining the others. 
For consistency and simplicity, I renamed the 'title' series.
I then needed the monetary columns in a float type. I stripped them from any non-numeric value ('$', ','), then converted them.
To make them more readable, I divided them by million.

```python
# prepare budget for join

budget= budget.rename(columns={'movie':'title'})
budget['domestic_gross'] = budget['domestic_gross'].apply(lambda a: str(a).replace('$', '').replace(',','')).astype(float)
budget['production_budget'] = budget['production_budget'].apply(lambda a: str(a).replace('$', '').replace(',','')).astype(float)
budget['worldwide_gross'] = budget['worldwide_gross'].apply(lambda a: str(a).replace('$', '').replace(',','')).astype(float)
#divide by million 
budget['domestic_gross'] = budget['domestic_gross']/1000000
budget['production_budget']= budget['production_budget']/1000000
budget['worldwide_gross'] = budget['worldwide_gross']/1000000
budget.info()
budget.head()
```

I merged 'Movies' and 'Budget' by 'title' column.

```python
# Join movies with budget

movies =budget.merge(movies, on='title', how='left')
movies .head()
```

Clean the data.

All data were in a single dataframe, but it still needed some polishing.
I dropped non-numeric char from the 'foreign_gross' column and converted it into a float. The merging process created two similar and unnecessary columns: 'domestic_gross_x' and 'domestic_gross_y'. I calculated their mean and stored the result in a new column called 'domestic_gross'.
'Domestic_gross' and 'foreign_gross' have been divided by million, so their values are easier to read.
I dropped several columns as not needed. The resulting dataframe is much cleaner and maintainable.  

```python
# Clean data 

um = movies # ultimate movies

um.dropna(inplace=True) # remove null rows

#convert object to float
um['foreign_gross'] = um['foreign_gross'].apply(lambda a: str(a).replace(',','')).astype(float)

#calculate domestic gross mean
um['domestic_gross'] = (um['domestic_gross_x'] + um['domestic_gross_y']) / 2   # TODO find a better solution. THIS IS VERY GROSS

#drop unwanted column 
um = um.drop(['id','release_date','title','tconst','original_title','start_year','studio','domestic_gross_x','domestic_gross_y','year'], axis=1)

#divide by milion
um['domestic_gross'] = um['domestic_gross']/1000000
um['foreign_gross'] = um['foreign_gross']/1000000

um.info()
um.head()
```

Sorted value by worldwide_gross. Movies with the highest revenues are on the top.

```python
#sort value um by worldwide_gross

um=um.sort_values(by='worldwide_gross' ,ascending=False)
um
```

The values in the 'genre' column were often formed of three words. They needed to be more specific and easier to aggregate. I decided to keep only the first definition, dropping all that was coming after. As a result, the number of values shrunk from 192 to 16 genres. 

```python
# Reduce genres to a more meaninful amount

dummies = um['genres'].str.split(',') #splits the genre on the comma and returns a list
dummies = dummies.str[0] #select only the first element on the list
um= um.drop(['genres'], axis=1)
um['genres'] = dummies
um.info()
um.head()

```

I further filter the 'genre' column. I counted each genre frequency, keeping only those with at least 50 instances. 
I then created a new dataframe called 'Umg', grouping the row by the identified genre.
Data were ready for analysis. 

```python
# Group the rows by genre
count = um.groupby(['genres'])['genres'].count()
filt_count = count.loc[count > 50]
g = filt_count.index.values
filt_um = um.loc[um['genres'].isin(g)]
ugm=filt_um.groupby(['genres']).mean() #ultimate grouped movies
#um=um.first()  Select initial periods of time series data based

ugm.info()
ugm
```

I divided 'numvotes' column by 100.

```python
#divide numvotes by 100

ugm['numvotes'] = ugm['numvotes']/100

ugm 
```

I added a new column call 'profits', calculated it by subtracting 'production_budget' from 'worldwide_gross'.

```python
# add a column

ugm ['profits']= ugm ['worldwide_gross'] - ugm['production_budget']
ugm
```

I reorganised the dataframe sorting by 'worldwide_gross' and by 'profits'

```python
#sort value ugm by worldwide_gross

ugm_W=ugm.sort_values(by='worldwide_gross' ,ascending=False)
ugm_W
```

```python
#sort value ugm by profits

ugm_P=ugm.sort_values(by='profits' ,ascending=False)
ugm_P
```

```python
#reset index

ugm_P.reset_index (inplace=True)
```
